```python
import torch
from diffusers import StableDiffusionPipeline

# 1. Choose the Model ID (v1.5 is standard, stable, and fast)
model_id = "runwayml/stable-diffusion-v1-5"

# 2. Load the Pipeline
# 'torch_dtype=torch.float16' helps it run faster and fit in Colab memory
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# 3. Connect the Pipeline to your GPU (the 'cuda' part)
pipe = pipe.to("cuda")

print("The assembly line is ready!")    
```

```python
# 1. Define the Prompt (What you WANT to see)
# Tip: Use descriptive adjectives and style keywords
prompt = "create a full moon, cinematic lighting, hyper-realistic, 8k "

# 2. Define the Negative Prompt (What you DON'T want to see)
# This helps clean up common AI mistakes like blurriness or distortion
n_prompt = "blurry, low quality, distorted, extra fingers, text, watermark"

# 3. Generate the Image
# 'guidance_scale' tells the AI how strictly to follow your text (7.5 is sweet spot)
image = pipe(prompt, negative_prompt=n_prompt, guidance_scale=7.5).images[0]

# 4. Show and Save
image.save("ai_lab_result.png")
image
```
